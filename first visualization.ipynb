```python
import pandas as pd 
import numpy as np 

data=pd.read_csv("healthcare-dataset-stroke-data.csv")

data = data.dropna(axis=0)
```

```python
bmi=data["bmi"].values.tolist()
ht=data["hypertension"].values.tolist()
age=data["age"].values.tolist()

stroke=data["stroke"].values.tolist()

gender=data["gender"].values.tolist()

heart_disease=data["heart_disease"].values.tolist()
```

```python

mbmi=[]
fbmi=[]
mage=[]
fage=[]
mht=[]
fht=[]
mstroke=[]
fstroke=[]
mheart_disease=[]
fheart_disease=[]
for i in range(len(gender)):
    if gender[i]=="Male":
        mbmi.append(bmi[i])
        mage.append(age[i])
        mht.append(ht[i])
        mstroke.append(stroke[i])
        mheart_disease.append(heart_disease[i])
    else:
        fbmi.append(bmi[i])
        fage.append(age[i])
        fht.append(ht[i])   
        fstroke.append(stroke[i])
        fheart_disease.append(heart_disease[i])
```

```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
# Data for three-dimensional scattered points
ms=[1*(mheart_disease[n]+1) for n in range(len(mbmi))]
fs=[2*(fheart_disease[n]+1) for n in range(len(fbmi))]
print()
ax.scatter3D(mbmi, mage, mht,s=ms,c=mstroke,cmap='bwr')
ax.scatter3D(fbmi, fage, fht,s=fs,c=fstroke,cmap='bwr',marker="^")
fig.set_size_inches(18.5, 10.5)
```
